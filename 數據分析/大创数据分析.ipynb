```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

from wordcloud import WordCloud, STOPWORDS

# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Read the whole text.
text = open(path.join(d, 'alice.txt')).read()

# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open(path.join(d, "alice_mask.png")))

stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue',font_path='./fonts/simhei.ttf')

# generate word cloud
wc.generate(text)

# store to file
wc.to_file(path.join(d, "alice.png"))

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
fx = df[['19、您的性别','20、您的年龄层（文化程度）？','21、您的月均消费水平是？']]
fx.columns = ['性别','年龄层（文化程度）','月均消费水平']
```

```python
fx.describe()
```

```python
def getNameVal(gfx):
    name = []
    val = []
    for na,df in list(gfx):
        name.append(na)
        val.append(df.shape[0])
    return name,val
```

```python
f1 = fx.groupby('性别')
f2 = fx.groupby('年龄层（文化程度）')
f3 = fx.groupby('月均消费水平')
```

```python
name,val = getNameVal(f1)
from pyecharts import Bar
bar = Bar("性别分布")
bar.add("性别分布",name, val,is_more_utils=True)
bar
```

```python
name,val = getNameVal(f2)
from pyecharts import Bar
bar = Bar('年龄层（文化程度）')
bar.add('年龄层（文化程度）',name, val,is_more_utils=True)
bar
```

```python
name,val = getNameVal(f3)
from pyecharts import Bar
bar = Bar('月均消费水平')
bar.add('月均消费水平',name, val,is_more_utils=True)
bar
```

```python
fx2 = df[['4、您对网红产品一般会持怎样的态度？','19、您的性别','20、您的年龄层（文化程度）？','21、您的月均消费水平是？']]
fx2.columns = ['态度','性别','年龄层（文化程度）','月均消费水平']
f1 = fx2.groupby('态度')
```

```python
Name,val = getNameVal(f1)
from pyecharts import Bar
bar = Bar("好感度分布情况")
bar.add("好感度分布情况",Name, val,is_more_utils=True)
bar
```

```python
fx2 = fx2.replace({'不感兴趣':60,'怀疑':70,'中立':80,'好奇且保持新鲜感':90,'热衷':100})
f2 = fx2.groupby('性别')
f3 = fx2.groupby('年龄层（文化程度）')
f4 = fx2.groupby('月均消费水平')
```

```python
list(f2)
def my_draw(bar, f2, Name, NAME):
    name = []
    val = []
    for na,df in list(f2):
        name.append(na)
        a,b,c,d,e=0,0,0,0,0
        for i in range(0,df.shape[0]):
            if df.iloc[i,0]==60:
                a += 1
            elif df.iloc[i,0]==70:
                b += 1
            elif df.iloc[i,0]==80:
                c += 1
            elif df.iloc[i,0]==90:
                d += 1
            elif df.iloc[i,0]==100:
                e += 1
        val.append([a,b,c,d,e])
    for i in range(0,len(name)):
        print(name[i],val[i])
    bar = Bar(NAME)
    for i in range(0,len(name)):
        bar.add(name[i], Name, val[i], is_stack=True)
    return bar
```

```python
my_draw(bar, f2, Name, "好感度-性别")
```

```python
my_draw(bar, f3, Name, "年龄层-好感度")
```

```python
my_draw(bar, f4, Name, "月均消费-好感度")
```
